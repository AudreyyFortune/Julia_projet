## 4) Comparer 2 séquences

Nous allons prendre comme exemple ici deux fichiers issues de la Protein Data Bank : 6HQ1 et 1UHM qui sont respectivement des domaines globulaires de l'histone H1 humaine, et de l'homologue d'histone de liaison Hho1p de *S. cerevisiae*. 


```julia
# import des packages

Pkg.add("BioAlignments")
using BioAlignments
```

```julia
# domaine globulaire de l'histone humaine H1.0
# domaine globulaire de l'homologue d'histone de liaison Hho1p de S. cerevisiae
struc1, struc2 = retrievepdb.(["6HQ1", "1UHM"])
```

```julia
seq1, seq2 = LongAminoAcidSeq.([struc1["A"], struc2["A"]], standardselector, gaps=false)
println("La séquence protéique du fichier '6HQ1.pdb' est : $seq1")
println()
println("La séquence protéique du fichier '1UHM.pdb' est : $seq2")
```

### * Matrice de score Blosum62

```julia
scoremodel = AffineGapScoreModel(BLOSUM62, gap_open = -10, gap_extend = -1)
# Une matrice de substitution est une fonction du score de substitution (ou du coût) d'un symbole à l'autre. 
```

### * Alignement global

```julia
alignment_global = pairalign(GlobalAlignment(), seq1, seq2, scoremodel)
```

```julia
printstyled("Concernant l'alignement global, nous avons: ", color=:blue, '\n')
println("score = ",score(alignment_global)) # Score
aln_global=alignment(alignment_global)
println("nombre de positions matchées = ",count_matches(aln_global)) #Count the number of matching positions.
println("nombre de positions mismatchées = ",count_mismatches(aln_global)) # Count the number of mismatching positions.
println("nombre insertions = ",count_insertions(aln_global)) #Count the number of inserting positions.
println("nombre délétions = ",count_deletions(aln_global)) #Count the number of deleting positions.
println("nombre de positions alignées = ",count_aligned(aln_global)) # Count the number of aligned positions.
```

### *  Alignement local

```julia
alignment_local = pairalign(LocalAlignment(), seq1, seq2, scoremodel)
```

```julia
printstyled("Concernant l'alignement local, nous avons: ", color = :blue, '\n')
println("score = ",score(alignment_local)) # Score
aln_local=alignment(alignment_local)
println("nombre de positions matchées = ",count_matches(aln_local)) #Count the number of matching positions.
println("nombre de positions mismatchées = ",count_mismatches(aln_local)) # Count the number of mismatching positions.
println("nombre insertions = ",count_insertions(aln_local)) #Count the number of inserting positions.
println("nombre délétions = ",count_deletions(aln_local)) #Count the number of deleting positions.
println("nombre de positions alignées = ",count_aligned(aln_local)) # Count the number of aligned positions.
```

### * RMSD

```julia
r = rmsd(struc1, struc2)
println("Le rmsd de l'alignement '6HQ1.pdb' et '1UHM.pdb' est de : $r")
```
